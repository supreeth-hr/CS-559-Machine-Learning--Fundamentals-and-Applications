```python
#Importing the required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
```

```python
# Loading the test and train data
test_data = pd.read_csv("test.csv")
train_data = pd.read_csv("train.csv")
```

```python
train_data
```

```python
test_data=test_data.drop("ID",axis=1)
test_data
```

```python
# Separate features and labels of both the test and train data
X_test = test_data.iloc[:, :-1].values  # Features
y_test = test_data.iloc[:, 3].values   # True labels

X_train = train_data.iloc[:, :-1].values  # Features
y_train = train_data.iloc[:, 3].values   # True labels
```

```python
#Standardizing the data
scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

Answer: 1) Classifying using 3-nearest neighbors

```python
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

print("Predicted values for y_test: \n", y_pred)
print("\nActual values for y_test: \n", y_test)
print("\nConfusion matrix for the above prediction: \n", confusion_matrix(y_test, y_pred))
print("\nClassification report the above prediction: \n", classification_report(y_test,y_pred)) 
```

Answer 2) Classifying using Euclidean distance weighted 3-nearest neighbors

```python
classifier_weighted = KNeighborsClassifier(n_neighbors=3, metric="euclidean",weights="distance")
classifier_weighted.fit(X_train, y_train)
y_pred_weighted = classifier_weighted.predict(X_test)

print("Predicted values for y_test: \n", y_pred_weighted)
print("\nActual values for y_test: \n", y_test)
print("\nConfusion matrix for the above prediction: \n", confusion_matrix(y_test, y_pred_weighted))
print("\nClassification report the above prediction: \n", classification_report(y_test,y_pred_weighted)) 
```

From above we can see that, both "3-nearest neighbors" and "weighted 3-
nearest neighbors" classify the test data with the same results. Hence, for the given input (data and the value of k) both the models are giving the results. However, if we increase the amount of data and change the value of "K", we might see that, "weighted 3- nearest neighbors" will work better.
